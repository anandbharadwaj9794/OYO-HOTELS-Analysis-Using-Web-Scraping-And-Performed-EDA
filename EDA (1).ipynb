# Univariate Analysis - Distribution of Hotels by City

```python
city_counts = df['City'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(city_counts.values, labels=city_counts.index, autopct='%1.1f%%', startangle=120, colors=plt.cm.Set2.colors)
plt.title("Distribution of Hotels by City")
plt.axis('equal')  
plt.show()
```

# Insights:
>More number of Hotels in Hyderabad which is around 37% hotels across these 6 cities.
>
>Less number of Hotels in Goa which is around 4.3% hotels across these 6 cities.
>
>Delhi and Kolkata has almost equal number of hotels which is around 19%

# Bivariate Analysis - Average Hotel Rating by City

```python
plt.figure(figsize=(8,5))
city_rating = df.groupby('City')['Clean_Rating'].mean().reset_index().sort_values(by='Clean_Rating', ascending=False)
sns.barplot(x='City', y='Clean_Rating', data=city_rating, palette='magma')
plt.title('Average Hotel Rating by City', fontsize=14)
plt.xlabel('City')
plt.ylabel('Average Clean_Rating')
plt.show()

```

# Insights:
All cities maintain high cleanliness standards, with average ratings above 4

# Bivariate Analysis - Relationship between Rating and Price

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Clean_Rating', y='Clean_Price', data=df, hue='Price_Category', palette='coolwarm')
plt.title('Relationship between Rating and Price', fontsize=14)
plt.xlabel('Clean_Rating')
plt.ylabel('Clean_Price')
plt.show()
```

# Insights:
> Budget Hotels at lower price points, showing minimal variation in pricing regardless of cleanliness rating.
>
> Mid-range Hotels span a broader range of ratings and prices, indicating a balance between quality and cost.
>
> Premium hotels are more spread across higher ratings

# Multivaraite Analysis

# Multivaraite Analysis - Average Price and Rating by City & Price Category

```python
plt.figure(figsize=(10,6))
avg_stats = df.groupby(['City', 'Price_Category'])[['Clean_Price', 'Clean_Rating']].mean().reset_index()

sns.barplot(x='City', y='Clean_Price', hue='Price_Category', data=avg_stats, palette='Set2')
plt.title('Average Price by City and Price Category', fontsize=14)
plt.xlabel('City')
plt.ylabel('Average Clean_Price')
plt.legend(title='Price_Category')
plt.show()
```

# Insights:
> Budget hotels are consistent pricing across cities which means a competitive environment in these segment.
>
>Mid-range prices vary more noticeably, with Goa and Hyderabad trending higher.
>
> Premium listings command the highest prices in every city, with Goa leading significantly


# Multivaraite Analysis - Correlation Heatmap

```python
plt.figure(figsize=(8,6))
corr = df[['Clean_Price', 'Clean_Rating', 'Has_AC', 'Has_Wifi', 'Has_TV']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numeric Variables', fontsize=14)
plt.show()
```

# Insights:
>Price vs Rating â†’ 0.08 Very weak positive correlation Suggests that higher prices donâ€™t necessarily guarantee better cleanliness ratings. Guests may not perceive value purely through price Mumbai stands out in the Premium category,
>
>Has AC vs Has TV â†’ 0.69 Strong positive correlation Indicates that AC and TV often co-occur, likely  within mid-range or premium hotels.
